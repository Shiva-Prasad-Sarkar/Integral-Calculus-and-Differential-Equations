```python
# 1
from sympy import *
x = symbols('x')
f = (x**2 - 7*x + 6) / (x - 10)
f1 = diff(f, x)
f2 = diff(f, x, 2)
cp = solve(f1, x)
for i in cp:
  if i==0:
    print('NO MAX OR MIN VALUE')
    break
  else:
    if f2.subs(x,i)<0:
      max = f.subs(x,i)
      print('The maximum value is =',max,'at point',i)
    else:
      min = f.subs(x,i)
      print('The minimum value is =',min,'at point',i)
```

```python
# 2
from sympy import *
x = symbols('x')
f = x/ln(x)
f2 = diff(f, x)
f3 = diff(f2,x)
cp = solve(f2, x)
for i in cp:
  if f3.subs(x, i) < 0:
    max = f.subs(x,i)
    print('The maximum value is : ', max)
  else:
    print(f"The function has no maxima")
```

```python
# 3
from sympy import *

x = symbols('x')
y = x**2 * ln(x)

y3 = diff(y, x, 3)

result = y3.subs(x, 2)

print(f'Third derivative of y = x^2 * ln(x): {y3}')
print(f'Third derivative at x = 2: {result}')
```

```python
# 4 (a)
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
x,a,b = smp.symbols('x,a,b')
f = x**3-3*x+2
f1 = smp.diff(f,x)
f2 = smp.diff(f1,x)
f3 = smp.diff(f2,x)
plt.title('Plot of Functions')
d = np.arange(-5, 5, 0.1)
f = smp.lambdify(x,f)
f1 = smp.lambdify(x,f1)
f2 = smp.lambdify(x,f2)
f3 = smp.lambdify(x,f3)
plt.axvline(x=0, c='gray')
plt.axhline(y=0, c='gray',linestyle=':')
plt.plot(d, f(d), c='red', linestyle='dashed', label='f(x)')
plt.plot(d, f1(d), c='green', linestyle='-', label='f1(x)')
plt.plot(d, f2(d), c='blue', linestyle='-.', label='f2(x)')
plt.plot(f3(d), c='black', linestyle=':', label='f3(x)')
plt.ylim(-5, 5)
plt.xlim(-10,10)
plt.xlabel('x-axis',c='blue')
plt.ylabel('y-axis',c='red')
plt.legend(loc='upper right')
plt.show()
```

```python
# 4 (b)
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
x,a,b = smp.symbols('x,a,b')

f = a*smp.sin(3*x)
eq = f.subs(a, 5)

f1 = smp.diff(eq,x)
f2 = smp.diff(f1,x)
f3 = smp.diff(f2,x)
plt.title('Plot of Functions')
d = np.arange(-5, 5, 0.5)
eq = smp.lambdify(x,eq)
f1 = smp.lambdify(x,f1)
f2 = smp.lambdify(x,f2)
f3 = smp.lambdify(x,f3)
plt.axvline(x=0, c='gray')
plt.axhline(y=0, c='gray',linestyle=':')
plt.plot(d, eq(d), c='red', linestyle='dashed', label='f(x)')
plt.plot(d, f1(d), c='green', linestyle='-', label='f1(x)')
plt.plot(d, f2(d), c='blue', linestyle='-.', label='f2(x)')
plt.plot(d,f3(d), c='black', linestyle=':', label='f3(x)')
plt.ylim(-5, 5)
plt.xlim(-10,10)
plt.xlabel('x-axis',c='blue')
plt.ylabel('y-axis',c='red')
plt.legend(loc='upper right')
plt.show()
```

```python
# 5
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2*np.pi, 200)
x = 5 * np.cos(theta)
y = 10 * np.sin(theta)
fig, ax = plt.subplots(1)
ax.plot(x, y)
x2 = x * np.cos(np.pi/4) - y * np.sin(np.pi/4)
y2 = x * np.sin(np.pi/4) + y * np.cos(np.pi/4)
ax.plot(x2, y2)
x3 = x * np.cos(np.pi/2) - y * np.sin(np.pi/2)
y3 = x * np.sin(np.pi/2) + y * np.cos(np.pi/2)
ax.plot(x3, y3)
x4 = x * np.cos(3*np.pi/4) - y * np.sin(3*np.pi/4)
y4 = x * np.sin(3*np.pi/4) + y * np.cos(3*np.pi/4)
ax.plot(x4, y4)

ax.set_aspect('equal')
plt.title('Atom')
plt.show()
```
